```python
import optuna

def objective(trial):
    # The following suggest method specifies integer points between -5 and 5.
    x = trial.suggest_float("x", -5, 5, step=1)
    return x**2

# Non-int points are specified in the grid.
search_space = {"x": [-0.4, 0.5, 25]}
study = optuna.create_study(direction="maximize",sampler=optuna.samplers.GridSampler(search_space))
study.optimize(objective, n_trials=2)
```

```python
import numpy as np
import pandas as pd

def log_int(x, base : int=2):
  return np.floor(np.log(x)/np.log(base)).astype(int)


def generate_sample_numbers(y, base : int, n_rungs : int) -> list[int]:
  
  data_size = len(y)
  data_scale = log_int(data_size, base)
  min_scale = data_scale - n_rungs
  min_samples = base**min_scale
  
  return [
      *map(lambda scale: base**scale, range(min_scale, data_scale+1))
  ]
```

```python
random_series = pd.Series(np.random.rand(129))

n_samples_list = generate_sample_numbers(random_series, 2, 5)

for n_samples in n_samples_list:
    print(f"Number of samples: {n_samples}")
    X_train_sample = random_series.sample(n_samples, random_state=42)
```

```python
import inspect

def filter_args(func: callable, **kwargs):
    sig = inspect.signature(func)
    return {k: v for k, v in kwargs.items() if k in sig.parameters}

y = pd.Series(np.random.rand(129))

params = {
    "base": 2,
    "n_rungs": 3,
    "tomatos": 5,
}

generate_sample_numbers(y, **filter_args(generate_sample_numbers, **params))
```
