```python
import numpy as np
import pandas as pd
from pathlib import Path

data = Path(r"C:/Users/nikol/Documents/GitHub/PhytovirusDB/data/features")

# Load the feature DataFrames
feature_files = {
    "ctd": "X_ctd.tsv",
    "ctdc": "X_ctdc.tsv",
    "ctdt": "X_ctdt.tsv",
    "ctdd": "X_ctdd.tsv",
    "aac": "X_aac.tsv",
    "fb2b": "X_b2btools.tsv",
    "fnsp": "X_netsurfp.tsv",
    "faa": "X_residue.tsv",
    "fbiophys": "X_biophysical.tsv",
    "faaclass": "X_class.tsv",
}

all_X = {
    name: pd.read_csv(data / fname, sep="\t", index_col="Prot_ID")
    for name, fname in feature_files.items()
}
```

```python
# read each X in all_x and search for all na values append column of each na value with count of na values to a list and print the list

na_counts = []
for name, df in all_X.items():
    na_count = df.isna().sum().sum()
    na_counts.append((name, na_count))

for name, count in na_counts:
    print(f"{name}: {count} NA values")
```

```python
# read each X in all_x and search for all na values by column and print the count of na values for each column
for name, df in all_X.items():
    na_counts = df.isna().sum()
    na_counts = na_counts[na_counts > 0].sort_values(ascending=False)
    print(f"{name}:")
    print(na_counts)
    print("\n")
```
